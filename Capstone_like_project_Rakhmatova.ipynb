##Downloading data

```python
import pandas as pd

URL = " https://noaaocm.blob.core.windows.net/ais/csv2/csv2025/ais-2025-01-08.csv.zst"

df = pd.read_csv(URL)
```

```python
df
```

##Set the index to the base_date_time

```python
df['base_date_time'] = pd.to_datetime(df['base_date_time'])
df = df.set_index('base_date_time').sort_index()
```

```python
df.loc["2025-01-08 00:00:00"]
```

##Clean the data

```python
df[df['sog'].isnull()]
```

```python
df = df.dropna(subset=['sog'])
```

```python
df['sog'].isnull().sum()
```

```python
df
```

##First Analysis: When are boats out? Display a line graph where your x-axis is the hour of the day, and the y-axis is the number of unique vessels out at that hour.

```python
import matplotlib.pyplot as plt

hourly_vessels = df.groupby(df.index.hour)['mmsi'].nunique()
hourly_vessels.plot(kind='line')

plt.xlabel('Hour of Day')
plt.ylabel('Number of Unique Vessels')
plt.title('Number of Vessels at Sea by Hour')
```

###The highest number of vessels is around 18–19 UTC, while the lowest is around 09 UTC. This generally makes sense, as more activity happens during daytime hours. However, since the data includes vessels from different time zones, activity stays fairly steady throughout the day rather than showing a clear day–night pattern.

##Second Analysis: let's look at a specific boat, the Henry Hudson (MMSI: 366651000). The owner of the boat has asked us how long the boat engines were active for that day.

###a) Figure out when the boat was "accelerating". Create a column called "acceleration", which is going to be the difference between each "sog"

```python
henry = df[df['mmsi'] == 366651000]
```

```python
henry = henry.sort_index()
```

```python
henry['acceleration'] = henry['sog'].diff()
henry[['sog', 'acceleration']].head()
```

###b) Create a duration column, which is the difference between each "base_date_time"

```python
henry['duration'] = henry.index.to_series().diff()
```

```python
henry.head()
```

###c) Once you have a column "acceleration" and "duration", we can assume that the boat engines were active if the "acceleration" is greater than 0. Filter your dataframe to only have rows where the acceleration is positive.

```python
henry_active = henry[henry['acceleration'] > 0]
```

```python
henry_active[['sog', 'acceleration', 'duration']].head()
```

###d) Sum up the "duration" column to see how long the engines were active for that day. This will allow us to answer the boat captain

```python
henry_active_time = henry_active['duration'].sum()
print(henry_active_time)
```

####The total engine active time for the vessel Henry Hudson is 08:03:24 (approximately 8 hours)
